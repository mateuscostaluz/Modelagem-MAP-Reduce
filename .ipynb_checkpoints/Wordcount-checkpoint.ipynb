```python
import sys
import re
from functools import reduce
```

```python
def mapper(nome):
    arquivo = open(nome, 'r', encoding='utf-8-sig')
    linhas = map(lambda linha: re.sub('[^A-Za-z0-9]+', ' ', linha).lower().split(), arquivo)
    palavras = reduce(lambda x, y: x + y, linhas)
    tuplas = map(lambda palavra: (palavra, 1), palavras)
    return list(tuplas)
```

```python
def partitioner(lista):
    listaTuplasParticao = list(map(lambda linha: (linha[0], [linha[1]]), lista))
    dicionario = {}
    for palavra in lista:
        if palavra[0] in dicionario:
            dicionario[palavra[0]].append(palavra[1])
        else:
            dicionario[palavra[0]] = [palavra[1]]
    listaRepeticoesPalavras = []
    for repeticao in dicionario.items():
        listaRepeticoesPalavras.append(repeticao)
    return listaRepeticoesPalavras
```

```python
def reducer(lista):
    return list(map(lambda tupla: (tupla[0], reduce(lambda x, y: x + y, tupla[1])), lista))
```

```python
def odernaMaioesrOcorrencias(listaOcorrencia):
    return(sorted(listaOcorrencia, key = lambda x: x[1], reverse = True))
```

```python
nome = 'Beyond Good and Evil - Friedrich Nietzsche.txt'
```

```python
listaTuplas = mapper(nome)
```

```python
listaTuplas
```

```python
listaTuplasParticao = partitioner(listaTuplas)
```

```python
listaTuplasParticao
```

```python
listaTuplasOcorrencia = reducer(listaTuplasParticao)
```

```python
listaTuplasOcorrencia
```

```python
listaTuplasOcorrenciaOrdenada = odernaMaioesrOcorrencias(listaTuplasOcorrencia)
```
